```python
bm25 = "nq-dev.json"
reranked_bm2 = "nq-dev-plm-T0_3B-topk-1000.json"
wikipedia_file = "psgs_w100.tsv"
```

```python
from dpr_wiki_dataset import OpenRetrievalEvidenceDataset
```

```python
dataset = OpenRetrievalEvidenceDataset('2018 Wikipedia from DPR codebase',
                                       'evidence', 
                                       wikipedia_file,
                                       None)
```

```python
!head nq-dev.json
```

```python
import json
```

```python
data = json.load(open(bm25))
```

```python
data_reranked = json.load(open(reranked_bm2))
```

```python
filtered_data = []
for item in data_reranked:
    question = item['question']
    first_passage_reranked = item['ctxs'][0]
    answer = item['answers']
    for item2 in data:
        if question == item2['question']:
            first_passage_bm25 = item2['ctxs'][0]
            filtered_data.append({'question': question, 
                                  'answer': answer, 
                                  'original': first_passage_bm25, 
                                  'reranked': first_passage_reranked})
```

```python
from pprint import pprint
for item in filtered_data:
    if (item['original']['has_answer'] == True) and (item['reranked']['has_answer'] == False):
        print(item['question'])
        print(item['answer'])
        print("Original:\t", dataset.id2text[item['original']['id']])
        print("Reranked:\t", dataset.id2text[item['reranked']['id']])
        print()
        print()
```
